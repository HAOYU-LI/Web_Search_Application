```python
def match(pattern, string):
    lenPattern = len(pattern)
    lenString = len(string)
    
    if(lenString < lenPattern):
        return False
    
    result = False
    count = 0
    for char in string:
        if char == pattern[count]:
            count += 1
        elif char != pattern[count]:
            if char == pattern[0]:
                count = 1
            else:
                count = 0
        
        if count == lenPattern:
            result = True
            break
            
    return result
    
    
```

```python
def find_pdf_urls(page):
    pattern = 'content_cvpr_2018/papers/'
    pdf_urls = list()
    for ele in page:
        if match(pattern, ele):
            paperName = re.search('href=(.+)>pdf', ele).group(1)
            paperLink = 'http://openaccess.thecvf.com/' + paperName[1:len(paperName) - 1]
            pdf_urls.append(paperLink)
    return pdf_urls
```

```python
def download_file(urls, save_path = '/Users/lihaoyu/Documents/USC/INF551/project/Web_Search_Application/data/' + '2018'):
    if urls is None or len(urls) == 0:
        return
    
    if not os.path.isdir(save_path):
        os.mkdir(save_path)
        
    for i in tqdm(range(len(urls))):
        file_name = re.search('papers/(.+)',urls[i]).group(1)
        url_content = requests.get(urls[i], stream = True)
        pdf_path = os.path.join(save_path, file_name)
        
        with open(pdf_path, 'wb') as pdf:
            for chunk in url_content.iter_content(chunk_size = 1024):
                if chunk:
                    pdf.write(chunk)
```

```python
from PyPDF2 import PdfFileReader
import os
```

```python
def get_info(path):
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f)
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()

    #print(info)
    author = info.author
    creator = info.creator
    producer = info.producer
    subject = info.subject
    title = info.title
    
    return info, number_of_pages
```
